```python
import glob
import os

# Get list of all *_data.csv files in /content/
csv_files = glob.glob("/content/*_data.csv")

# Create an empty list to store DataFrames
dfs = []

for file in csv_files:
    # Extract the ticker from the filename
    ticker = os.path.basename(file).split('_')[0]

    # Read the CSV
    df = pd.read_csv(file)

    # Drop the first two rows (metadata rows)
    df = df.drop(index=[0, 1]).reset_index(drop=True)

    # Rename 'Price' to 'Date' if needed
    df = df.rename(columns={'Price': 'Date'})

    # Keep only the required columns
    df = df[['Date', 'Close', 'High', 'Low', 'Open', 'Volume']]

    # Add a 'Symbol' column with the ticker name
    df['Symbol'] = ticker

    # Append cleaned df to list
    dfs.append(df)

# Combine all into a single DataFrame
combined_df = pd.concat(dfs, ignore_index=True)

# Optional: Save the combined dataframe to CSV
#combined_df.to_csv('/content/sp500_combined_2025.csv', index=False)

# Show the first few rows
combined_df.head()
```

```python
combined_df = combined_df.sort_values(by=['Symbol', 'Date']).reset_index(drop=True)
```

### SP500 Company Profile

```python
import yfinance as yf
import pandas as pd
from tqdm import tqdm  # for progress bar

# Step 1: Load the S&P 500 tickers
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
tables = pd.read_html(url)
sp500_df = tables[0]
tickers = sp500_df['Symbol'].tolist()

# Step 2: Create a list to store company profile data
company_data = []

# Step 3: Download profile info for each ticker
for ticker in tqdm(tickers):
    try:
        info = yf.Ticker(ticker).info
        company_data.append({
            "Symbol": ticker,
            "Shortname": info.get("shortName", None),
            "Sector": info.get("sector", None),
            "Industry": info.get("industry", None),
            "Currentprice": info.get("currentPrice", None),
            "Marketcap": info.get("marketCap", None)
        })
    except Exception as e:
        print(f"Failed for {ticker}: {e}")

# Step 4: Convert to DataFrame
profile_df = pd.DataFrame(company_data)

# Step 5: Optional  Estimate Weight as Marketcap % of total
total_marketcap = profile_df["Marketcap"].sum(skipna=True)
profile_df["Weight"] = profile_df["Marketcap"] / total_marketcap * 100
```

```python
import yfinance as yf
import pandas as pd
from datetime import datetime

# Sector ETFs
sector_etfs = {
    "Basic Materials": "XLB",
    "Communication Services": "XLC",
    "Consumer Cyclical": "XLY",
    "Consumer Defensive": "XLP",
    "Energy": "XLE",
    "Financial Services": "XLF",
    "Healthcare": "XLV",
    "Industrials": "XLI",
    "Real Estate": "XLRE",
    "Technology": "XLK",
    "Utilities": "XLU"

}

# Date range
start_date = "2025-01-02"
end_date = "2025-04-03"

results = []

for sector, ticker in sector_etfs.items():
    data = yf.download(ticker, start=start_date, end="2025-04-04")
    try:
        jan2_close = data.loc["2025-01-02"]["Close"]
        apr3_close = data.loc["2025-04-03"]["Close"]
        ytd_change = ((apr3_close - jan2_close) / jan2_close) * 100
        results.append({"Sector": sector, "YTD % Change": round(ytd_change, 2)})
    except KeyError:
        print(f"Data not available for {sector} on required dates.")

# Save or display
df = pd.DataFrame(results)
print(df)
df.to_csv("sector_etf_YTD_2025.csv", index=False)
```

```python
# Read in the file (which may be a stringified series instead of a real CSV)
with open("sector_etf_YTD_2025.csv", "r") as f:
    raw_text = f.read()

# Clean the raw string
import re
lines = raw_text.strip().split('\n')
data = []
for line in lines:
    match = re.match(r"([A-Z]+)\s+([-\d.]+)", line)
    if match:
        ticker, ytd = match.groups()
        data.append((ticker, float(ytd)))

# Convert to DataFrame
import pandas as pd
df_ytd = pd.DataFrame(data, columns=["Ticker", "YTD % Change"])
print(df_ytd.head())
```

```python
import yfinance as yf
import pandas as pd
from datetime import datetime

# List of S&P 500 industries (we'll use SPDR sector ETFs as proxies)
industry_etfs = {
    "Advertising Agencies": "XLC",  # Using Communication Services as proxy
    "Aerospace & Defense": "XAR",
    "Airlines": "JETS",
    "Banks": "KBE",
    "Biotechnology": "XBI",
    "Chemicals": "XLB",
    "Consumer Electronics": "XLY",
    "Gold": "GLD",
    "Healthcare": "XLV",
    "Oil & Gas": "XOP",
    "Real Estate": "XLRE",
    "Retail": "XRT",
    "Semiconductors": "SOXX",
    "Software": "IGV",
    "Technology": "XLK",
    "Utilities": "XLU"
    # Add more mappings as needed
}

# Get current date (April 3, 2025 in your case)
end_date = '2025-04-03'
start_date = '2025-01-01'

# Fetch data and calculate YTD returns
ytd_returns = {}

for industry, ticker in industry_etfs.items():
    try:
        data = yf.download(ticker, start=start_date, end=end_date)
        if not data.empty:
            start_price = data['Adj Close'].iloc[0]
            end_price = data['Adj Close'].iloc[-1]
            ytd_return = (end_price - start_price) / start_price * 100
            ytd_returns[industry] = round(ytd_return, 2)
    except Exception as e:
        print(f"Error processing {industry}: {str(e)}")

# Convert to DataFrame and display
results = pd.DataFrame.from_dict(ytd_returns, orient='index', columns=['YTD Return (%)'])
print(results.sort_values('YTD Return (%)', ascending=False))
```
