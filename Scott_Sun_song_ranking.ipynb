```python
import pandas as pd
df_daily_ranking = pd.read_csv('data.csv')
df_daily_ranking.head()
```

```python
df_daily_ranking[df_daily_ranking.isnull().any(axis=1)].head()
```

```python
df_daily_ranking.isnull().any(axis=1)
```

```python
df_daily_ranking.dropna(inplace=True)
```

Part 3: Analyze the dataset

```python
#filter song in the USA
df_daily_ranking_us = df_daily_ranking[df_daily_ranking['Region'] == 'us']
df_daily_ranking_us.head()
```

```python
#rank the artists in usa
df_daily_ranking_us.groupby('Artist').sum()
```

```python
df_daily_ranking_us.groupby('Artist').sum().sort_values('Streams', ascending=False)['Streams'].head(n=5)
```

```python
df_daily_ranking_us.groupby('Track Name').sum().sort_values('Streams', ascending=False)['Streams'].head(n=5)
```

```python
artist = df_daily_ranking_us.groupby('Artist').sum().sort_values('Streams', ascending=False).head(n=5).index
```

```python
artist
```

```python
df_artist = df_daily_ranking_us[df_daily_ranking_us['Artist'].isin(artist)]
df_artist
```

```python
df_artist.groupby(['Date', 'Artist']).sum()
```

```python
#reset_index pulls out the attributes used for groupby groupby back to normal columns
df_chart = df_artist.groupby(['Date', 'Artist']).sum()['Streams'].reset_index()
df_chart.head(10)
```

```python
import altair as alt
```

```python
#alt.data_transformers.disable_max_rows will disable the 5000 rows limitation set by altair.
alt.data_transformers.disable_max_rows()
```

```python
alt.Chart(df_chart).mark_line().encode(
    x='Date:T',
    y='Streams:Q',
    color='Artist:N'
)
```

```python
tracks = df_daily_ranking_us.groupby('Track Name').sum().sort_values('Streams', ascending=False).head(n=5).index
```

```python
tracks
```

```python
df_tracks = df_daily_ranking_us[df_daily_ranking_us['Track Name'].isin(tracks)]
```

```python
df_tracks
```

```python
df_tracks_2 = df_tracks.groupby(['Date', 'Track Name']).sum()['Streams'].reset_index()
```

```python
df_tracks_2
```

```python
alt.Chart(df_tracks_2).mark_line().encode(
    x='Date:T',
    y='Streams:Q',
    color='Track Name:N'
)
```
